```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import display, HTML

def calculate_similarity(text1, text2):
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([text1, text2])
    similarity = cosine_similarity(vectors[0], vectors[1])[0][0]
    return similarity

def detect_plagiarism(original_text, suspicious_text, threshold=0.8):
    similarity_score = calculate_similarity(original_text, suspicious_text)

    if similarity_score >= threshold:
        return True
    else:
        return False

# Example usage:
original_content = "This is the original content created by the author."
suspicious_content = "This is some content that may have been copied from the original."

is_plagiarized = detect_plagiarism(original_content, suspicious_content)

output = "Plagiarism detected!" if is_plagiarized else "Life is all about doing your best in trying to find what works out for you and taking most time in trying to pursue those skills."
display(HTML(output))
```
